```python
optab = pd.read_csv('Optab.txt')
```

```python
# Create an empty dataframe
Machine_code = pd.DataFrame(columns=['Machine_code'])
```

```python
lst = optab.iloc[:, 0].tolist()

a = 0
for i in ass_text.iloc[:, 1]:
  if i == 'HLT':
      Machine_code = Machine_code.append({'Machine_code': 5000}, ignore_index=True)
      break
  elif i in lst:
    val = str(optab.loc[optab['Mnemonic'] == i,'Opcode'].values[0]) + '00' + str(ass_text.iloc[a, 0])
    Machine_code = Machine_code.append({'Machine_code': val}, ignore_index=True)
    a+=1
  else:
    Machine_code = Machine_code.append({'Machine_code': 0}, ignore_index=True)
    a+=1
```
